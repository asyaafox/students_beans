# Рост учеников и бобиков

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
bob=pd.read_csv('fasol.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame named 'data'

# Create the first displot with hue (separate lines for each gender)
sns.displot(
    data[''],
    x='рост',
    bins=15,
    kde=True
)
plt.ylabel("Количество")
plt.show()
```

```python
sns.displot(data[(data['класс']=='10c')],x='рост',binwidth=2,kde=True)
```

```python
sns.displot(data,x='рост',binwidth=2,kde=True)
```

```python
# Create the first displot with hue (separate lines for each gender)
sns.displot(
    bob,
    x='размер боба мм',
    bins=14,
    kde=True
)
plt.ylabel("Количество")
plt.show()
```
